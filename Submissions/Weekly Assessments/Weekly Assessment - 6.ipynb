```python
#1. Write a recursive function for the Fibonacci sequence.

def fibonacci(n):
    start = 0
    if n <= 0:
        return start
    elif n == 1:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)
fibonacci(10)
```

```python
#2. Explain "Memoization" and how it helps recursive functions.

# Memoization is an optimization technique used to speed up recursive functions by storing the results of expensive function calls 
# and returning the cached result when the same inputs occur again. In the context of recursive functions, especially those that 
# have overlapping subproblems like the Fibonacci sequence, memoization can significantly reduce the number of calculations needed.
```

```python
#3. Use reduce() from functools to find the product of a list.

from functools import reduce
numbers = [1, 2, 3, 4, 5]
product = reduce(lambda x, y: x * y, numbers)
print(product)
```

```python
#4. Write a function to check if a number is an "Armstrong Number."

def armstrong(n):
    num_str = str(n)
    digits = len(num_str)
    sum_of_powers = sum(int(digit) ** digits for digit in num_str)
    return sum_of_powers == n

print(armstrong(153))  
print(armstrong(123))
```

```python
# 5. Create a closure that remembers a base value and adds it to an input.
def create_adder(base):
    def adder(x):
        return x + base
    return adder

add_five = create_adder(5)
print(add_five(10))  
```

```python
# Explain the "LEGB" rule in Python scope.

# LEGB stands for Local, Enclosing, Global, Built-in.
# It is the order in which Python looks up variable names when they are referenced.
# 1. Local (L): The innermost scope, which is the local scope of the current function.
# 2. Enclosing (E): for nested loops, etc.
# 3. Global (G): accessible from anywhere after the variable is defined.
# 4. Built-in (B): The outermost scope, which contains the keywords and built-in names of Python.
```

```python
# 7. Write a function that converts a decimal number to binary without bin().

def binary(n):
    if n == 0:
        return "0"
    binary_str = ""
    while n > 0:
        binary_str = str(n % 2) + binary_str
        n //= 2
    return binary_str
```

```python
# 8. Create a "Generator" function using yield to produce infinite even numbers.

def generate():
    n = 0
    while True:
        yield n
        n += 2

# main
generator = generate()
for _ in range(5):
    print(next(generator))
```

```python
# 10. Write a function to validate a password (min 8 chars, 1 digit, 1 symbol).

def validate_password(password):
    if len(password) < 8:
        return False
    has_digit = any(char.isdigit() for char in password)
    has_symbol = any(not char.isalnum() for char in password)
    return has_digit and has_symbol

validate_password("Sherwin_24")
```
