```python
num1 = 10
num2 = 5
operation = "+"

if operation == "+":
    result = num1 + num2
elif operation == "-":
    result = num1 - num2
elif operation == "*":
    result = num1 * num2
elif operation == "/":
    result = num1 / num2 if num2 != 0 else "Cannot divide by zero"
else:
    result = "Invalid operation"

print(f"Result: {result}")
```

```python
numbers = [4, 12, 8, 23, 7]

largest = numbers[0]
for num in numbers:
    if num > largest:
        largest = num

print(f"The largest number is: {largest}")
```
