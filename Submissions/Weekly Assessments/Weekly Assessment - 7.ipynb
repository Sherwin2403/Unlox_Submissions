```python
# 1. Define a class Car with attributes and a method to display details.

class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def display_details(self):
        print(f"Car Details: {self.year} {self.make} {self.model}")

car1 = Car("Toyota", "Camry", 2020)
car1.display_details()
```

```python
# 3. Create a subclass ElectricCar that inherits from Car.
# 4. Demonstrate "Method Overriding" in inheritance. (both in the same cell)
class ElectricCar(Car):
    def __init__(self, make, model, year, battery_size):
        super().__init__(make, model, year) # inherting the parent class' attributes
        self.battery_size = battery_size

    def display_details(self):
        super().display_details() # inherting the parent class' method
        print(f"Battery Size: {self.battery_size} kWh") 

electric_car = ElectricCar("Tesla", "Model 3", 2021, 100)   
electric_car.display_details()
```

```python
# 5. Explain "Encapsulation" by creating private attributes using __.
class Car:
    def __init__(self, make, model, year):
        self.__make = make # private attribute using __ between '.' and the attribute name
        self.__model = model
        self.__year = year

    def display_details(self):
        print(f"Year: {self.__year}, Make: {self.__make}, Model: {self.__model}")
```

```python
# 6. Use @property to create getter and setter methods.

class Car: 
    def __init__(self, make, model, year):
        self.__make = make
        self.__model = model
        self.__year = year

    @property
    def make(self):
        return self.__make

    @make.setter
    def make(self, value):
        self.__make = value

    @property
    def model(self):
        return self.__model

    @model.setter
    def model(self, value):
        self.__model = value

    @property
    def year(self):
        return self.__year

    @year.setter
    def year(self, value):
        self.__year = value

#Example usage of the Car class with getter and setter methods

car = Car("Toyota", "Camry", 2020)
print(car.make)  # Output: Toyota
car.make = "Honda"
print(car.make)  # Output: Honda    
```

```python
# 7. What is "Multiple Inheritance"? Provide a small code structure.

# feature where a class inherits attributes and methods from more than one parent class.

# Example:

class Engine:
    def start_engine(self):
        print('Engine started')

class Wheels:
    def rotate_wheels(self):
        print('Wheels are rotating')

class Car(Engine, Wheels):
    def drive(self):
        self.start_engine()
        self.rotate_wheels()
        print('Car is driving')

car = Car()
car.drive()
```

```python
# 8. Create a class method using the @classmethod decorator.

class Car:
    def __init__(self, make, model):
        self.make = make
        self.model = model

    @classmethod 
    def from_string(cls, car_string): 
        make, model = car_string.split('-')
        return cls(make, model)
    
car_string = "Toyota-Camry"
car = Car.from_string(car_string)
print(car.make)  
print(car.model)  
```

```python
# 9. Explain the difference between a class method and a static method.

# A class method is a method that is bound to the class and not the instance of the class. 
# It takes the class as its first argument (conventionally named 'cls'). 
# Class methods can access and modify the class state.

# A static method is a method that is bound to the class and not the instance of the class. 
# It does not take the class or instance as its first argument. 
# Static methods cannot access or modify the class state.

# Example of class method and static method:

class MathUtils:
    @classmethod
    def add(cls, a, b):
        return a + b

    @staticmethod
    def multiply(a, b):
        return a * b
```
