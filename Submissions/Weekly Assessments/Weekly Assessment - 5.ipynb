```python
# A function that accepts variable-length arguments (*args).

def variable_length_args(*args):
    if not args:
        return "No arguments were provided."
    
    arg_list = []
    
    for arg in args:
        arg_list.append(str(arg))
    
    result = ", ".join(arg_list)
    
    return f"Arguments received: {result}"

list = ['sherwin', 'jaysingh', 'shaunwin', 'pratapsingh']
variable_length_args(*list)
```

```python
#Create a function that accepts keyword arguments (kwargs) and prints them.

def print_kwargs(**kwargs):
    if not kwargs:
        return "No keyword arguments were provided."
    
    result = []
    
    for key, value in kwargs.items():
        result.append(f"{key}: {value}")
    
    return "Keyword arguments received:\n" + "\n".join(result)

print(print_kwargs(name="Sherwin", age=25, city="New York"))
```

```python
# 3)Write a recursive function to calculate the factorial of a number

def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers."
    elif n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
    
print("Factorial of 4:", factorial(4))
print("Factorial of 8:", factorial(8))
print("Factorial of 5:", factorial(5))
```

```python
# 4)Explain the difference between Global and Local scope with a code example.

'''Global scope refers to variables that are defined outside of any function 
and can be accessed from anywhere in the code. Local scope refers to variables 
that are defined within a function and can only be accessed within that function.'''

#example of Global and Local scope:

global_var = "Sherwin" # This is a global variable

def example_function():
    local_var = "John" # This is a local variable
    print(global_var) 
    print(local_var) 

example_function() # will print both global and local variables
print(global_var) # Accessing global variable

# print(local_var) # This will cause an error because local_var is not defined in this scope
```

```python
# 5)Write a lambda function to sort a list of tuples based on the second element.

def sort_tuples(tuples_list):
    sorted_list = sorted(tuples_list, key=lambda x: x[1])
    return sorted_list
 
list = [(1, 'b'), (2, 'a'), (3, 'c')]
sorted_result = sort_tuples(list)
print(sorted_result)
```

```python
# Create a "Filter" function to extract prime numbers from a list

def filter_primes(numbers):
    def is_prime(n):
        if n <= 1:
            return False
        for i in range(2, int(n**0.5 + 1)):
            if n % i == 0:
                return False
        return True

    prime_list = []
    for n in numbers:
        if is_prime(n):
            prime_list.append(n)
    return prime_list

numbers = [2, 3, 4, 5, 6, 7, 8, 9, 10]
prime_numbers = filter_primes(numbers)
print("Prime numbers:", prime_numbers)
```

```python
# 9)Demonstrate the use of nonlocal in a nested function

def outer_function():
    x = 10
    def inner_function():
        nonlocal x
        x = 20
    inner_function()
    return x

print(outer_function())

''' when a variable is declared as nonlocal, it means that the variable is not local to the inner function, 
but it is also not global. It allows the inner function to modify the variable in the outer function's scope.'''
```

```python
# 10)Create a function decorator that prints "Starting..." before any function runs

def start_decorator(func):
    def wrapper(*args, **kwargs):
        print("Starting...")
        return func(*args, **kwargs)
    return wrapper

def function_to_decorate():
    print("Function is running.")

decorated_function = start_decorator(function_to_decorate)
decorated_function()

```
