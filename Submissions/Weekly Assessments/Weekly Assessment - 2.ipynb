```python
# 4) Factorial Computation

def fact(n):
  product = 1
  while n > 1 :
    product = product * n
    n = n - 1
  return product

#main
n = int(input("Enter a number : "))
product = fact(n)
print(f"Factorial of {n} is : {product}")
```

```python
# 5) The Fibonacci Series:

def fibonacci(n):
  array_series = []
  for i in range(n):
    if i == 0:
      array_series.append(0)
    elif i == 1:
      array_series.append(1)
    else:
      array_series.append(array_series[i - 2] + array_series[i - 1])
  print(array_series)

n = int(input("Enter a number : "))
fibonacci(n)
```

```python
# Section - 2
# 6) Prime  Number Validation
import math as m
def is_prime(n):
  if n == 1 or n == 0:
    print(f"{n} is not prime")
  elif m.sqrt(n) == 1:
    print(f"{n} is prime")
  else:
    count = 0
    for i in range(2, m.floor(m.sqrt(n)) + 1):
      if n % i == 0:
        count = count + 1
    if count > 0:
      print(f"{n} is not prime")
    else:
      print(f"{n} is prime")

#main function
n = int(input("Enter a number : "))
is_prime(n)

```

```python
# 7) Integer Reversal
def reverse(n):
  reversed = 0
  copy = n
  while (copy > 0):
    remainder = copy % 10
    reversed = reversed * 10 + remainder
    copy = copy//10
  return reversed


n = int(input("Enter the number : "))
result = reverse(n)
print(f"Reverse of {n} is {result}")

```

```python
# 8) Numerical Palindrome
def reverse(n):
  reversed = 0
  copy = n
  while (copy > 0):
    remainder = copy % 10
    reversed = reversed * 10 + remainder
    copy = copy//10
  return reversed


n = int(input("Enter the number : "))
result = reverse(n)
print(f"Reverse of {n} is {result}")
if result == n:
  print(f"{n} is a palindrome")
else:
  print(f"{n} is not a palindrome")
```

```python
# 9) Amstrong number

def amstrong(n):
  copy = n
  sum = 0
  while copy > 0:
    remainder = copy % 10
    sum = sum + remainder ** 3
    copy = copy // 10
  return sum == n

n = int(input("Enter the number : "))
result = amstrong(n)
if result:
  print(f"{n} is an amstrong number")
else:
  print(f"{n} is not an amstrong number")
```

```python
# 10) Euclidean Algorithm

def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a
n1 = int(input("Enter the first number : "))
n2 = int(input("Enter the second number : "))
result = gcd(n1, n2)
print(f"The greatest common divisor is {result}")
```
